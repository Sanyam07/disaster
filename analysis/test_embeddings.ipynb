```python
path = '{0}/glove6B/glove.6B.300d.txt'.format(config.download_path)

f = open(path,'r')

glove = {}

for line in f:

    splitLine = line.split()
    word = splitLine[0]
    embedding = np.array([float(val) for val in splitLine[1:]])
    glove[word] = embedding
```

```python
lst = "['the', 'forest', 'is', 'on', 'fire', 'what', 'a', 'disaster']"

np.average([glove[s.lower()] for s in literal_eval(lst) if s.lower() in glove.keys()], axis=0)
```

```python
dic = {
    '0': "['the', 'forest', 'is', 'on', 'fire', 'what', 'a', 'disaster']",
    '1': "['the', 'forest', 'is', 'nice']",
    '2': "['vsf']",
    '3': "[]"
}

df = pd.DataFrame(dic.items(), columns=['id', 'tokens'])

df
```

```python
mask = [True, False, False, False]

literal_eval(df[mask]['tokens'].values[0])
```

```python
mask = [True, False, False, False]

np.average([
    glove[s] for s in 
    literal_eval(df[mask]['tokens'].values[0])
    if s in glove.keys()
],
axis=0)
```

```python
df['embeddings_array'] = df.apply(
        lambda row: (
            np.average(
                [
                    glove[s] for s in 
                    literal_eval(row['tokens'])
                    if s in glove.keys()
                ],
                axis=0
            )
        ),
        axis=1
    )

df['embeddings_array']
```

```python
# Drop nan
df.dropna(axis=0, subset=['embeddings_array'], inplace=True)

df['embeddings_array']
```

```python
cols = ['feature_glove_{0}'.format(str(s)) for s in range(0, 300)]

df.set_index('id', inplace=True)

df[cols] = pd.DataFrame(
    df['embeddings_array'].values.tolist(),
    index=df.index
    )
```
