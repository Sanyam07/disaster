```python
from disaster import config  # noqa
import pandas as pd

for s in ['train', 'test']:

    # Load data

    path = (
        '{0}/nlp-getting-started/{1}_processed.csv'
        .format(config.data_path, s)
    )
    
    df = pd.read_csv(path)
    
df
```

```python
df['hash'] = df.apply(lambda row: row['hash'].replace('[]', '["none_hash"]'), axis=1)
df['at'] = df.apply(lambda row: row['at'].replace('[]', '["none_at"]'), axis=1)

display(df[['id', 'hash']].head(10))
display(df[['id', 'at']].head(10))
```

```python
from ast import literal_eval

df['hash'] = df.apply(lambda row: literal_eval(row['hash']), axis=1)
df['at'] = df.apply(lambda row: literal_eval(row['at']), axis=1)

display(df[['id', 'hash']].head(10))
display(df[['id', 'at']].head(10))
```

```python
df['array'] = df.apply(lambda row: list(set(row['hash'] + row['at'])), axis=1)

df[['id', 'array']].head(10)
```

```python
df_exploded = df.explode('array')[['id', 'array']].drop_duplicates().set_index('id')

df_exploded
```

```python
# Get one hot encoding of columns B
df_hot = (
    pd.get_dummies(df_exploded['array'])
    .groupby('id')
    .sum()
    .reset_index()
)

df_hot
```

```python
mask = (df_hot['earthquake'] == 1) & (df_hot['target_'] == 0)

df_hot[mask][['id', 'earthquake', 'text', 'target_']]
```

```python
mask = (df_hot['earthquake'] == 1) & (df_hot['target_'] == 1)

df_hot[mask][['id', 'earthquake', 'text', 'target_']]
```
